```python
from sklearn.cluster import KMeans
```

```python
# List to store inertia values for different k (number of clusters)
inertia = []
```

```python
X_train
```

```python
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_train)
    inertia.append(kmeans.inertia_)
```

```python
# Plot the Elbow Curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Within-Cluster Sum of Squared Distances)')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

The  value of k seems to be 2 as well as 3. But inorder to determine which is the best let us evaluate using 3 methods.

1. Silhouette score
We will calculate the silhouette score for both k=2 and k=3.
The silhouette score measures how similar an object is to its own cluster compared to other clusters.
A higher silhouette score indicates better-defined clusters.

```python
from sklearn.metrics import silhouette_score

# Fit KMeans for k=2
kmeans_2 = KMeans(n_clusters=2, random_state=42)
labels_2 = kmeans_2.fit_predict(X_train)
```

```python
# Fit KMeans for k=3
kmeans_3 = KMeans(n_clusters=3, random_state=42)
labels_3 = kmeans_3.fit_predict(X_train)
```

```python
# Calculating silhouette scores
silhouette_k2 = silhouette_score(X_train, labels_2)
silhouette_k3 = silhouette_score(X_train, labels_3)

print(f'Silhouette Score for k=2: {silhouette_k2}')
print(f'Silhouette Score for k=3: {silhouette_k3}')
```

The silhouette score for this is greater for k=2

2. Davies-Bouldin Index

This index evaluates the average similarity ratio of each cluster with its most similar cluster
. A lower Davies-Bouldin index indicates better clustering.

```python
from sklearn.metrics import davies_bouldin_score

db_index_k2 = davies_bouldin_score(X_train, labels_2)
db_index_k3 = davies_bouldin_score(X_train, labels_3)

print(f'Davies-Bouldin Index for k=2: {db_index_k2}')
print(f'Davies-Bouldin Index for k=3: {db_index_k3}')
```

The David-Bouldin Index also shows better resu;ts for k=2

The k-means++ initialization method is a smarter way to choose the initial centroids. It helps to spread out the initial centroids, which can lead to better convergence and improved clustering results compared to random initialization. The k-means++ method works as follows:

It selects the first centroid randomly from the data points.
For each subsequent centroid, it chooses a point from the dataset with a probability proportional to its squared distance from the nearest existing centroid. This helps to ensure that the centroids are well spread out.

```python
Kmeans=KMeans(n_clusters=3,init='k-means++')
Kmeans.fit(df)
```
